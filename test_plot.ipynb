```python
import numpy as np
import matplotlib.pyplot as plt
import os

# 读取数据的函数
def load_data(s, a, v, folder="data_EWC"):
    # 定义子文件夹路径
    subfolder = os.path.join(folder, f"s_{s} a_{a} v_{v}")
    
    # 检查文件是否存在并加载
    try:
        Error = np.load(os.path.join(subfolder, "E.npy"))
        Weights = np.load(os.path.join(subfolder, "W.npy"))
        Configuration = np.load(os.path.join(subfolder, "C.npy"))
        print(f"成功加载数据从文件夹: {subfolder}")
        return Error, Weights, Configuration
    except FileNotFoundError as e:
        print(f"文件未找到: {e}")
        return None, None, None

E, C, W = load_data(s=25, a=0.30, v=0.44)
print('C:',C.shape)
print(E.shape)
```

```python
n_steps = 30000
dt = 0.001
nn_dim = 18
```

```python
dpi = 100
width = 960
height = 640
figsize = (width / dpi, height / dpi)
timevals = np.linspace(0,n_steps*dt,n_steps)
_, ax = plt.subplots(figsize=figsize, dpi=dpi, sharex=True)

ax.plot(timevals, C[:,1])
ax.set_ylabel('m')
ax.set_title('Deepth of net')
```

```python
dpi = 100
width = 960
height = 640
figsize = (width / dpi, height / dpi)
timevals = np.linspace(0,n_steps*dt,n_steps)
_, ax = plt.subplots(2, 2, figsize=figsize, dpi=dpi, sharex=True)

for i in range(nn_dim):
    ax[0,0].plot(timevals, W[0,:,i],label=f'w{i+1}')

ax[0,0].legend()
ax[0,0].set_title('Weights of MU 1')

for i in range(nn_dim):
    ax[0,1].plot(timevals, W[1,:,i])
ax[0,1].set_title('Weights of MU 2')

for i in range(nn_dim):
    ax[1,0].plot(timevals, W[2,:,i])
ax[1,0].set_title('Weights of MU 3')

for i in range(nn_dim):
    ax[1,1].plot(timevals, W[3,:,i])
ax[1,1].set_title('Weights of MU 4')
```
